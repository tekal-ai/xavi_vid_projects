```python
df_performance = df_performance[['ad_id', 'asset_id', 'carrousel_position', 'spend', 'impressions', 'clicks', 'video_views', 'filename', 's3_url']]
```

```python
df_performance['ctr'] = df_performance.clicks / df_performance.impressions
df_performance['cpc'] = df_performance.spend / df_performance.clicks
df_performance['vtr'] = df_performance.video_views / df_performance.impressions

df_performance.loc[df_performance['clicks'] == 0, 'cpc'] = 0
df_performance.loc[df_performance['impressions'] == 0, 'ctr'] = 0
df_performance.loc[df_performance['impressions'] == 0, 'vtr'] = 0

df_performance['ctr'] = df_performance['ctr']*100
df_performance['vtr'] = df_performance['vtr']*100
```

```python
df_performance.head(3)
```

```python
df_annotation_1 = df_annotation.loc[df_annotation.asset_id.isin(df_performance.asset_id)]

df_performance_1 = df_performance.loc[df_performance.asset_id.isin(df_annotation.asset_id)]
```

```python
df_annotation_1.asset_id.nunique(), df_performance_1.asset_id.nunique()
```

```python
df_annotation_performance = pd.merge(df_annotation_1, df_performance_1, on='asset_id', how='outer')

df_annotation_performance
```

```python
import math
for i,e in enumerate(df_annotation_performance.start_frame):
    if math.isnan(e):
        df_annotation_performance.loc[i, 'logo_time'] = 'No data'
```
