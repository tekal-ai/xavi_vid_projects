```python
import numpy as np
import pandas as pd
pd.pandas.set_option('display.max_columns', None)
```

```python
df_hc = pd.read_csv('data/hair_care/HAIRCARE-ASSETS.csv')


df_hc.head()
```

```python
len(df_hc), len(df_hc.assetId.unique())
```

```python
df_hc_tags = pd.read_csv('data/hair_care/1212-BEAUTY-TAGS-ARG.csv')

df_hc_tags.head()
```

```python
df_hc = df_hc[['assetId', 'assetUrl', 'fileType', 'filename',
       'brand', 'brandId', 'sector', 'industry', 'adrecallScore',
       'adrecallProbability', 'impactScore', 'bai',
       'textSaliencyScore', 'fiveSecondIndex']]
```

```python
df_hc_tags = df_hc_tags[['id', 'name', 'assetId', 'tagType',
       'confidence', 'detectionId', 'startTimestamp', 'endTimestamp',
       'startFrame', 'endFrame']]
```

```python
df_hc_merged = pd.merge(df_hc_tags, df_hc, how='left', on='assetId')
```

```python
df_hc_merged = df_hc_merged.rename(columns={
    'id':'tag_id',
    'name':'tag_name',
    'assetId':'asset_id',
    'tagType':'tag_type',
    'brandId':'brand_id'
}).dropna(subset='brand_id')
```

```python
df_hc_merged.info()
```

```python
df_hc_merged.to_csv('data/hair_care/hair_care_tags_with_metrics.csv', index=False)
```
